```python
# Converting strings to epoch time
import pprint

time_list = table1["Time"]
date_list = table1["Date"]

h_array = []
m_array = []
d_array = []
epoch_array = []

for t in time_list:
  a = t.split(":")
  hour = int(a[0])
  b = a[1]
  if len(a) == 2 and b[-2:] == "pm" and int(a[0]) != 12:
      hour += 12
  else:
    if b[-1:] == "pm":
      hour += 12

  min = int(b[0:2])
  h_array.append(hour)
  m_array.append(min)
  # print(f"{hour:02}:{min:02}")

for d in date_list:
  # print(d[8::])
  d_array.append(int(d[8::]))

for i, c in enumerate(time_list):
   t = datetime(2023, 8, d_array[i], h_array[i], m_array[i], 0)
   epoch_array.append(int(calendar.timegm(t.timetuple())))

print(epoch_array)

# Converting tide heights to float numbers

height_list = table1["Feet"]
f_array = []

for h in height_list:
  f_array.append(float(h[0:4]))

print(f_array)
```

### Using sine to find side measures

```python
from math import nan, isnan, degrees, asin, acos, atan

angle = nan
opposite = 5
hypotenuse = 10

if isnan(opposite) is True:
  unknown = "opposite side"
  answer = hypotenuse * sin(radians(angle))
  print(f"The {unknown} is {round(answer, 2)} cm")

elif isnan(hypotenuse) is True:
  unknown = "hypotenuse"
  answer = opposite / sin(radians(angle))
  print(f"The {unknown} is {round(answer, 2)} cm")

elif isnan(angle) is True:
  unknown = "angle"
  answer = degrees(asin(opposite / hypotenuse))
  print(f"The {unknown} is {round(answer, 2)} degrees")

else:
  print("One of 'angle', 'opposite' and 'hypotenuse' must be nan")
```

### Using cosine to find side measures

```python
angle = 57
adjacent = 3
hypotenuse = nan

if isnan(adjacent) is True:
  unknown = "adjacent side"
  answer = hypotenuse * cos(radians(angle))
  print(f"The {unknown} is {round(answer, 2)} cm")

elif isnan(hypotenuse) is True:
  unknown = "hypotenuse"
  answer = adjacent / cos(radians(angle))
  print(f"The {unknown} is {round(answer, 2)} cm")

elif isnan(angle) is True:
  unknown = "angle"
  answer = degrees(asin(adjacent / hypotenuse))
  print(f"The {unknown} is {round(answer, 2)} degrees")

else:
  print("One of 'angle', 'adjacent' and 'hypotenuse' must be nan")
```

### Using tangent to find side measures

```python
angle = 57
opposite = nan
adjacent = 10


if isnan(adjacent) is True:
  unknown = "adjacent side"
  answer = opposite / tan(radians(angle))
  print(f"The {unknown} is {round(answer, 2)} cm")

elif isnan(opposite) is True:
  unknown = "opposite"
  answer = adjacent * tan(radians(angle))
  print(f"The {unknown} is {round(answer, 2)} cm")

elif isnan(angle) is True:
  unknown = "angle"
  answer = degrees(atan(opposite / adjacent))
  print(f"The {unknown} is {round(answer, 2)} degrees")

else:
  print("One of 'angle', 'adjacent' and 'hypotenuse' must be nan")
```

### The Unit Circle

All trig function graphs are derived from the Unit Circle

sin = y

cos = x

sin^2() + cos^2()  1 is thus derived via an application of Pythagoras' Theorem

IMG_1724.jpeg



IMG_1725.jpeg

```python
# Draw unit circle
fig, ax = plt.subplots()
plt.axis([-1.5, 1.5, -1.5, 1.5])
plt.plot([-2, 2], [0, 0], "b")
plt.plot([0, 0], [-2, 2], "b")
circle = plt.Circle((0, 0), 1, fill=False)
ax.add_artist(circle)
ax.set_aspect("equal")

# Inscribe right triangle inside unit circle
A = 1
x = cos(A)
y = sin(A)

plt.plot([x], [y], "r")
plt.plot([0, x], [0, y], "r")
plt.plot([0, x], [0, 0], "r")
plt.plot([x, x], [0, y], "r")
plt.plot([x], [y], "ro")

ax.text(0.2, -0.1, "x")
ax.text(0.6, 0.4, "y")
ax.text(0.05, 0.6, "r = 1")
ax.text(x + 0.1, y + 0.1, "(x, y)")

plt.show()
```

### Graph transformations

We'll model graph transformations using the sine graph but these principles apply to all trigonometric functions

```python
import matplotlib.pyplot as plt
import numpy as np

# Graph dimensions
xmin = -2
xmax = 14
ymin = -2
ymax = 2
points = 10 * (xmax - xmin)
x = np.linspace(xmin, xmax, points)

# Plot the axis
fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], "b")
plt.plot([0, 0], [ymin, ymax], "b")

y1 = np.sin(x)
plt.plot(x, y1, "r")

# Changes the amplitude by a factor of F
F = 0.5
y2 = F * np.sin(x)
plt.plot(x, y2, "g")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Graph dimensions
xmin = -2
xmax = 14
ymin = -2
ymax = 2
points = 10 * (xmax - xmin)
x = np.linspace(xmin, xmax, points)

# Plot the axis
fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], "b")
plt.plot([0, 0], [ymin, ymax], "b")

y1 = np.sin(x)
plt.plot(x, y1, "r")


# Changes the frequency by a factor of F
# The period and wavelength change
# Amplitude does not
F = 2
y2 = np.sin(F * x)
plt.plot(x, y2, "g")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Graph dimensions
xmin = -2
xmax = 14
ymin = -2
ymax = 2
points = 10 * (xmax - xmin)
x = np.linspace(xmin, xmax, points)

# Plot the axis
fig, ax = plt.subplots()
plt.axis([xmin, xmax, ymin, ymax])
plt.plot([xmin, xmax], [0, 0], "b")
plt.plot([0, 0], [ymin, ymax], "b")

y1 = np.sin(x)
plt.plot(x, y1, "r")


# Phase shift (horizontal shift)
# + shifts left
# - shifts right
F = -2
y2 = np.sin(x + F)
plt.plot(x, y2, "g")

plt.show()
```

> sin(x + /2) = cos(x) by the transformation above
